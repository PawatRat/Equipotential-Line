```python
def voltage_by_position(V,d,x):
    return V/d * (d-x)

def position_by_voltage(V_position,V=10,d=15):
    return d - ((V_position*d)/V)

def position_by_voltage_2(V_position,V=10,d=15):
    return (d * V_position) / V

def expand_list(list_outside,count=28):
    result = []
    for number in list_outside:
        ls = []
        for j in range(count+1):
            ls.append(number)
        result.append(ls)
    return result
```

```python
def rmse_score(y_actual, y1):
    mse1 = np.mean((y_actual-y1)**2)
    rmse1 = np.sqrt(mse1)

    return rmse1
```

```python
def theory_data_compare_2(data,color='m',medthod=""):
    '''
    data  is dict only
    '''

    th_avg_position = []
    ex_avg_position = []

    deviation = []
    x_plot = range(0,28+1)
    x_plot_2 = range(0,28+1,4)
    fig, ax = plt.subplots(1, figsize=(15, 15))

    # Experiment plot
    for voltage,plotInY in data.items():
        ax.scatter(x_plot_2,plotInY, color=color)
        coef = np.polyfit(x_plot_2,plotInY,1)
        p = np.poly1d(coef)
        ax.plot(x_plot_2,p(x_plot_2),color+medthod)
        ax.text(x_plot_2[-1]+0.8, p(x_plot_2)[-1]-0.05,
                voltage ,
                fontsize=15,
                color=color)
        th_avg_position.append(sum(plotInY) / len(plotInY))

    # Theory plot
    for votage_notreal,plotInY in enumerate(all_list[list(data.keys())[0]-1: list(data.keys())[-1]]):
        ax.scatter(x_plot,plotInY , color="b")
        coef = np.polyfit(x_plot,plotInY,1)
        p = np.poly1d(coef)
        ax.plot(x_plot,p(x_plot),"b")
        ax.text(x_plot[0]-1, p(x_plot)[-1]-0.05,
                votage_notreal+list(data.keys())[0] ,
                fontsize=15,
                color='b')
        ex_avg_position.append(sum(plotInY) / len(plotInY))

    for i in range(len(th_avg_position)):
        deviation.append(th_avg_position[i] - ex_avg_position[i])

    print("All line in surface",deviation)
    print("Average no absolute",sum(deviation)/len(deviation))
    print("Average of absolute",sum(list(map(abs, deviation)))/len(deviation))

    plt.xlabel("length of graph paper")
    plt.ylabel("width of graph paper")
    #plt.legend()
    plt.grid(b=True, which='major', color='#666666', linestyle='-')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.show()
```

```python
def rmse_plot_cal(data, color='m',medthod=""):
    rmse_lit = []
    all_slope = []

    x_plot_2 = range(0,28+1,4)
    fig, ax = plt.subplots(1, figsize=(15, 15))

    for voltage,plotInY in data.items(): # plot in each line
        ax.scatter(x_plot_2,plotInY, color=color)
        coef = np.polyfit(x_plot_2,plotInY,1)
        p = np.poly1d(coef)
        ax.plot(x_plot_2,p(x_plot_2),color+medthod)
        ax.text(x_plot_2[-1]+0.8, p(x_plot_2)[-1]-0.05,
        voltage,
        fontsize=15,
        color=color)
        rmse_lit.append(rmse_score(np.array([num for num in p(x_plot_2)]), np.array(plotInY)) )
        all_slope.append((p(x_plot_2)[3]-p(x_plot_2)[1])/(x_plot_2[3]-x_plot_2[1]))

    print("R square >> ",rmse_lit)
    print("Average R square >> ",sum(rmse_lit)/len(rmse_lit))

    print("slope in each line >> ",all_slope)
    print("Average slope >> ",sum(all_slope)/len(all_slope))

    plt.xlabel("length of graph paper")
    plt.ylabel("width of graph paper")
    #plt.legend()
    plt.grid(b=True, which='major', color='#666666', linestyle='-')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.show()
```

```python
def theory_data_compare(data, Vstart,Vend,color='m',medthod=""):
    th_avg_position = []
    ex_avg_position = []

    deviation = []
    x_plot = range(0,28+1)
    x_plot_2 = range(0,28+1,4)
    fig, ax = plt.subplots(1, figsize=(15, 15))

    # Experiment plot
    for plotInY in data:
        ax.scatter(x_plot_2,plotInY, label=data.index(plotInY)+2, color=color)
        coef = np.polyfit(x_plot_2,plotInY,1)
        p = np.poly1d(coef)
        ax.plot(x_plot_2,p(x_plot_2),color+medthod)
        th_avg_position.append(sum(plotInY) / len(plotInY))

    # Theory plot
    for plotInY in all_list[Vstart-1:Vend]:
        ax.scatter(x_plot,plotInY, label=volt_list.index(plotInY[0]), color="b")
        coef = np.polyfit(x_plot,plotInY,1)
        p = np.poly1d(coef)
        ax.plot(x_plot,p(x_plot),"b")

        ex_avg_position.append(sum(plotInY) / len(plotInY))

    for i in range(len(th_avg_position)):
        deviation.append(th_avg_position[i] - ex_avg_position[i])

    print("All line in surface",deviation)
    print("Average no absolute",sum(deviation)/len(deviation))
    print("Average of absolute",sum(list(map(abs, deviation)))/len(deviation))

    plt.xlabel("length of graph paper")
    plt.ylabel("width of graph paper")
    #plt.legend()
    plt.grid(b=True, which='major', color='#666666', linestyle='-')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.show()
```
